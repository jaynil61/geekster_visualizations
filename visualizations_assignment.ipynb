### Importing required libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Datasets

```python
bac = pd.read_csv('BAC.csv')
bac.head()
```

```python
c = pd.read_csv('C.csv')
c.head()
```

```python
gs = pd.read_csv('GS.csv')
gs.head()
```

```python
jpm = pd.read_csv('JPM.csv')
jpm.head()
```

```python
ms = pd.read_csv('MS.csv')
ms.head()
```

```python
wfc = pd.read_csv('WFC.csv')
wfc.head()
```

# Plotting different charts

### 1. Scatter plot

```python
plt.figure(figsize=(6, 4))
plt.scatter(bac['Open'], bac['Close'])
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.title('BAC: Open vs Close Price')
plt.show()
```

### 2. Heatmap

```python
correlation_matrix = bac[['Open','Close', 'Adj Close']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('BAC: Correlation Heatmap')
plt.show()
```

### 3. Bar-chart

```python
average_closing_prices = {
    'BAC': bac['Close'].mean(),
    'C': c['Close'].mean(),
    'GS': gs['Close'].mean(),
    'JPM': jpm['Close'].mean(),
    'MS': ms['Close'].mean(),
    'WFC': wfc['Close'].mean()
}

banks = list(average_closing_prices.keys())
closing_prices = list(average_closing_prices.values())

plt.figure(figsize=(6, 3))
plt.bar(banks, closing_prices)
plt.xlabel('Banks')
plt.ylabel('Average Closing Price')
plt.title('Average Closing Price of Banks')
plt.show()
```

### 4. Line-chart

```python
plt.figure(figsize=(6, 4))
plt.plot(bac['Date'], bac['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('BAC: Closing Price over Time')
plt.grid(True)
plt.show()
```

### 5. Pie-chart

```python
average_opening_prices = {
    'BAC': bac['Open'].mean(),
    'C': c['Open'].mean(),
    'GS': gs['Open'].mean(),
    'JPM': jpm['Open'].mean(),
    'MS': ms['Open'].mean(),
    'WFC': wfc['Open'].mean()
}

banks = list(average_opening_prices.keys())
opening_prices = list(average_opening_prices.values())

plt.figure(figsize=(5, 5))
plt.pie(opening_prices, labels=banks, autopct='%1.1f%%')
plt.title('Proportion of Average Opening Prices for Each Bank')
plt.show()
```

### 6. Boxplot

```python
closing_prices_all = [bac['Close'], c['Close'], gs['Close'], jpm['Close'], ms['Close'], wfc['Close']]

plt.figure(figsize=(6, 4))
plt.boxplot(closing_prices_all, labels=['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC'])
plt.xlabel('Banks')
plt.ylabel('Closing Prices')
plt.title('Boxplot of Closing Prices for Different Banks')
plt.show()
```

### 7. Violinplot

```python
closing_prices_all = [bac['Close'], c['Close'], gs['Close'], jpm['Close'], ms['Close'], wfc['Close']]

plt.figure(figsize=(6, 4))
sns.violinplot(data=pd.DataFrame(closing_prices_all, index=['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']).T)
plt.xlabel('Banks')
plt.ylabel('Closing Prices')
plt.title('Violin Plot of Closing Prices for Different Banks')
plt.show()
```

### 8. Histogram

```python
plt.figure(figsize=(6, 4))
plt.hist(bac['Close'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Histogram of BAC Closing Prices')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(bac['Close'], bins=20, kde=True, color='blue', edgecolor='black')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.title('Histogram of BAC Closing Prices with KDE')
plt.show()
```
